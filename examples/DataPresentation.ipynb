### Import Modules

```python
import matplotlib.pyplot as plt 
import numpy as np
import pandas as pd
import seaborn as sns
import os
output_dir = './plots_data/'
os.makedirs(output_dir, exist_ok=True)
colors = sns.color_palette("colorblind")
```

### Comparison between DRL Agent and Heuristics

```python
df = pd.read_csv(r"C:\Users\wanha\OneDrive\Desktop\Explainability\examples\plots_compare\blocking_rates_10.csv")

X = [50, 100, 150, 200, 250, 300, 350, 400, 450, 500]

Y1 = list(df.iloc[:, 1])
Y2 = list(df.iloc[:, 2])

plt.figure()
plt.plot(X, Y1, '-x', label='KSPFF', color=colors[0])
plt.plot(X, Y2, '-x', label='AGENT', color=colors[1])
plt.xlabel("Traffic Load (Errlangs)")
plt.ylabel("Blocking Probability Comparison")
plt.legend()
plt.title(f"Blocking Probability COST239")
plt.yscale('log')
plt.grid(which='minor', linewidth=0.3)
plt.minorticks_on()
plt.grid(which='major', linewidth=1)
plt.savefig(output_dir + f'BlockingProbability.png')
plt.show()
```

```python
df1 = pd.read_csv(r"C:\Users\wanha\OneDrive\Desktop\Explainability\examples\plots_compare\rewards_10.csv")
df2 = pd.read_csv(r"C:\Users\wanha\OneDrive\Desktop\Explainability\examples\plots_students\rewards_10.csv")

X = [100, 200, 300, 400, 500]

Y1 = list(df1.iloc[1::2, 2])
Y2 = list(df2.iloc[:, 1])

difference = np.array(Y1) - np.array(Y2)
difference = abs(difference)
Accuracy = abs(1 - (difference)/Y1)*100

plt.figure()
plt.plot(X, Accuracy, '-x', label='Student Accuracy', color=colors[0])
plt.xlabel("Traffic Load (Errlangs)")
plt.ylabel("Accuracy (%)")
plt.legend(loc = 4)
plt.title(f"Accuracy of Student for Agent")
plt.ylim([90, 100.5])
plt.axhline(y = 100, color = 'k', linestyle = '--')
plt.axhline(y = 97.5, color = 'k', linestyle = '--')
plt.grid(which='minor', linewidth=0.3)
plt.minorticks_on()
plt.grid(which='major', linewidth=1)
plt.savefig(output_dir + f'Accuracy_Agent_Reward.png')
plt.show()
```
